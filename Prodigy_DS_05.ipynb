```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
for chunk in df:
    missing = chunk.isna().sum().sort_values(ascending = False) / len(chunk)
    missing[missing != 0].plot(kind = 'barh')
    break
```

```python
for chunk in df:
    corr=chunk.corr()
    heat=sns.heatmap(corr)
    break
```

```python
for chunk in df:
    cities=chunk.City.nunique()
    print(cities)
    break
```

```python
for chunk in df:
    cities_accidents_count = chunk.City.value_counts()
    print(cities_accidents_count[:25])
    break
```

# Time during which most of the accidents occur

```python
for chunk in df:
    chunk['Start_Time'] = pd.to_datetime(chunk.Start_Time)
    break
```

```python
sns.distplot(chunk.Start_Time.dt.hour, bins = 24, norm_hist = True, kde=False)
```

This graph shows that most of the accidents happens during the peak traffic hours when people are commuting to work
